```python

import json
import csv 
from bs4 import BeautifulSoup
import requests
url="https://www.baraasallout.com/test.html"
r=requests.get(url=url)
print(r.text)
soup = BeautifulSoup(r.text, 'html.parser')
```

#task1

```python
data= []
for i in soup.find_all(["h1","h2"]):
    data.append(["head",i.get_text().strip()])

for i in soup.find_all("p"):
    data.append(["pargraph",i.get_text().strip()])

for i in soup.findAll("li"):
    data.append(["list",i.get_text().strip()])
file="abdo12.csv"
with open(file , mode="w")as file:
    w=csv.writer(file)
    w.writerow(["type","content"])
    w.writerows(data)


```

#task2

```python
tables=soup.findAll("table")
head=[]
for i in soup.findAll("th"):
    head.append(i.get_text().strip())
print(head)
d=[]
for i in soup.find_all("td"):
    d.append(i.get_text().strip())
print(d)
rows=[]
step=len(head)
for i in range(0,len(d),step):
    row=d[i:step+i]
    rows.append(row)
rows
file="w.csv"
with open(file , mode="w", newline="")as file :
    w=csv.writer(file)
    w.writerow(head)
    w.writerows(rows)
```

#task3

```python
all=soup.find_all('div',class_='card')
data= []
for card in all:
    data =card.find('h3')
    if data:
        data.append(data.get_text())
prices=[]
for card in all:
    price = card.find('span',class_='price')
    if price:
        prices.append(price.get_text())
stocks= []
for card in all:
    stock = card.find('span',class_='stock')
    if stock:
        stocks.append(stock.get_text())
button_texts= []
for card in all:
    button= card.find('button')
    if button:
        button_texts.append(button.get_text())
product_data= []
for i in range(len(all)):
    product_data.append({
        "title": data[i] if i < len(data)else None,
        "price": prices[i] if i < len(prices)else None,
        "stock_availability": stocks[i] if i < len(stocks)else None,
        "button_text": button_texts[i] if i < len(button_texts)else None
    })

with open('information.json','w')as file:
    json.dump(product_data, file)
```

#task4

```python
form =soup.find('form')
data=[]
for input_tag in form.find_all('input'):
    name= input_tag.get('name')
    type= input_tag.get('type')
    value= input_tag.get('value')
    data.append({
        "name": name,
        "type":type,
        "default_value":value})
with open('Form.json', 'w')as file:
    json.dump(data, file)
```

#task5

```python
links= []
for a in soup.find_all('a',href=True):
    links.append(a['href'])
videos = []
for b in soup.find_all('iframe', src=True):
    videos.append(b['src'])
data = {
    "links": links,
    "videos": videos
}
with open('Links.json', 'w')as file:
    json.dump(data, file)
```

#task6

```python
products= []
for product in soup.find_all('div', class_='featured-product'):
    id= product.get('data-id')
    name= product.find('span', class_='name').get_text()
    price= product.find('span', class_='price').get_text()
    color= product.find('span', class_='colors').get_text()

    products.append({
        "id": id,
        "name": name,
        "price": price,
        "colors":color })
with open('Featured_Products.json', 'w')as file:
    json.dump(products, file)
```
