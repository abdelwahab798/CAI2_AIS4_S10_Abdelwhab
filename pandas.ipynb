```python
def dataframe():    

    data = {'Name': ['Alice' , 'bob', 'Said','MM'] ,
        'Age' : [25,60,10,27],
        'City': ['Cairo', 'Seuz' , 'Lisbon','Ismailia']
        }
    df = pd.DataFrame(data)
    df['Salary'] = [700,1000,250,2500]

    return df

def dataframe_null():    

    data = {'Name': [None , 'bob', 'Said','MM'] ,
        'Age' : [25,None,10,None],
        'City': ['Cairo',None , 'Lisbon','Ismailia']
        }
   
    dfnull['Salary'] = [700,1000,250,None]
    dfnull = pd.DataFrame(data)
    return dfnull

df      = dataframe()
dfnull  = dataframe_null()
```
