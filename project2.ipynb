```python
import csv

# كلاس الموظف:بيكون فيه البيانات للموظف زي الاسم،رقمه،الوظيفة،المرتب
class Employee:
    def __init__(self, name, id, position, salary):
        self.name = name
        self.id = id
        self.position = position
        self.salary = salary
 # فانكشن لتحديث بيانات الموظفت
    def update_emp(self,name,position,salary):
        if name:
            self.name= name
        if position:
            self.position= position
        if salary:
            self.salary= salary


# كلاس لإدارة الموظفين والتعامل مع الملف
class EmployeeManager:
    def __init__(self, file="employees.csv"):
        self.file = file
        self.employees= self.get_employees()

    def get_employees(self):
        employees = []
        try:
            with open(self.file, mode="r") as file:
                reader= csv.DictReader(file)
                for row in reader:
                    emp =Employee(
                        name=row["Name"],
                        id=int(row["Id"]),
                        position=row["Position"],
                        salary=float(row["Salary"]),)
                    employees.append(emp)
        except FileNotFoundError:
            print("The file does not exist")
        except ValueError as c:
            print(f"Error reading the file:{c}")
        return employees

    def save_employees(self):
        with open(self.file, mode="w", newline="") as file:
            fieldnames = ["Name", "Id", "Position", "Salary"]
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for emp in self.employees:
                writer.writerow(
                    {
                        "Name":emp.name,
                        "Id":emp.id,
                        "Position":emp.position,
                        "Salary":emp.salary,
                    })
    def add_employee(self, name, id, position, salary):
        # التحقق من صحة البيانات
        if not id.isdigit():
            print("Error ID must be an integer")
            return
        try:
            salary = float(salary)
            if salary <= 0:
                print("errorSalary must be a positive number")
                return
        except ValueError:
            print("errorSalary must be a number")
            return

        for emp in self.employees:
            if int(id) == emp.id:
                print("Error this Id is already exist")
                return

        # إضافة موظف 
        emp = Employee(name=name, id=int(id), position=position, salary=salary)
        self.employees.append(emp)
        self.save_employees()
        print("employee added successfully")

    def update_employee(self, id, name=None, position=None, salary=None):
        # التحقق من صحة البيانات
        if not id.isdigit():
            print("error iD must be an integer")
            return
        if salary:
            try:
                salary = float(salary)
                if salary <= 0:
                    print("Error salary must be a positive number")
                    return
            except ValueError:
                print("errorsalary must be a number")
                return

        # تحديث بيانات الموظف
        for emp in self.employees:
            if int(id) == emp.id:
                emp.update_emp(name=name, position=position, salary=salary)
                self.save_employees()
                print("employee updated successfully")
                return
        print("Error employee with this ID does not exist")

    def display_employees(self):
        if not self.employees:
            print("no employees exit")
            return
        for emp in self.employees:
            print(
                f"Name: {emp.name}, ID: {emp.id}, Position: {emp.position}, Salary: {emp.salary}"
            )

    def find_employee(self, id):
        if not id.isdigit():
            print("error id must be an integer")
            return
        for emp in self.employees:
            if int(id) == emp.id:
                print(
                    f"Name: {emp.name}, ID: {emp.id}, Position: {emp.position}, Salary: {emp.salary}"
                )
                return
        print("error employee with this id does not exist")

    def delete_employee(self, id):
        if not id.isdigit():
            print("error Id must be an integer")
            return
        for emp in self.employees:
            if int(id) == emp.id:
                self.employees.remove(emp)
                self.save_employees()
                print("employee deleted successfully")
                return
        print("Error employee with this ID does not exist")


#القايمة الي هيسخدمها المستخدم
def menu():
    manager = EmployeeManager()
    while True:
        print("\nMenu:")
        print("1. Add Employee")
        print("2. Update Employee")
        print("3. Display Employees")
        print("4. Find Employee")
        print("5. Delete Employee")
        print("6. Exit")
        choice = input("Choose an option: ")
        if choice== "1":
            name= input("Enter Name:")
            id= input("Enter ID:")
            position= input("Enter Position: ")
            salary= input("Enter Salary:")
            manager.add_employee(name, id, position, salary)
        elif choice== "2":
            id=input("Enter ID: ")
            name = input("Enter Name:")
            position= input("Enter Position: ")
            salary= input("Enter Salary:")
            manager.update_employee(id, name=name, position=position, salary=salary)
        elif choice == "3":
            manager.display_employees()
        elif choice== "4":
            id= input("Enter ID: ")
            manager.find_employee(id)
        elif choice == "5":
            id= input("Enter ID: ")
            manager.delete_employee(id)
        elif choice=="6":
            print("exit from program")
            break
        else:
            print("plesae select the vaild number ")


menu()
```
